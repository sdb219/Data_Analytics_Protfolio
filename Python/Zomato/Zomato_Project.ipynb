```python
#Finding the top 10 cities with most restaurants

import matplotlib.pyplot as plt
import seaborn as sns

top_cities=df['city'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_cities.values,y=top_cities.index, palette='viridis', hue=top_cities.index, dodge=False, legend=False)
plt.title('Top 10 cities with Most Restaurants', fontsize=16)
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
#Rating Distribution

plt.figure(figsize=(10,5))
sns.countplot(x='aggregate_rating', data=df, palette='coolwarm',hue='aggregate_rating',dodge=False,legend=False)
plt.title('Distribution of Aggregate Ratings', fontsize=16)
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.tight_layout()
plt.show()
```

```python
#Average Cost for Two by City

# Find the dominant price range for each city
dominant_price_range = df.groupby('city')['price_range'].agg(lambda x: x.mode()[0])

# Average cost per city
avg_cost_city = df.groupby('city')['average_cost_for_two'].mean()

# Combine into a single DataFrame
city_cost_df = pd.DataFrame({
    'avg_cost': avg_cost_city,
    'dominant_price_range': dominant_price_range
}).reset_index()

# Sort descending by average cost and pick top 10
top_cities_cost = city_cost_df.sort_values(by='avg_cost', ascending=False).head(10)

# Plot with hue and legend
plt.figure(figsize=(12, 6))
sns.barplot(x='avg_cost', y='city', data=top_cities_cost,
            hue='dominant_price_range', palette='viridis')

plt.title('Top 10 Cities by Average Cost for Two with Dominant Price Range', fontsize=16)
plt.xlabel('Average Cost for Two')
plt.ylabel('City')
plt.legend(title='Dominant Price Range', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Filter for India (country_code = 1)
india_df = df[df['country_code'] == 1]
# Count restaurants with table booking by city
table_booking_cities = india_df[india_df['has_table_booking'] == 'Yes']['city'].value_counts().head(10)

print(table_booking_cities)

top_cities = table_booking_cities.index.tolist()

# Filter for only these cities
india_top_cities_df = india_df[india_df['city'].isin(top_cities)]

# Aggregate counts of 'Yes' for table booking and online delivery per city
booking_delivery_counts = india_top_cities_df.groupby('city').agg({
    'has_table_booking': lambda x: (x == 'Yes').sum(),
    'has_online_delivery': lambda x: (x == 'Yes').sum()
}).reset_index()

# Melt the dataframe for seaborn
booking_delivery_melted = booking_delivery_counts.melt(id_vars='city', 
                                                      value_vars=['has_table_booking', 'has_online_delivery'], 
                                                      var_name='Service', value_name='Count')

# Rename for clarity
booking_delivery_melted['Service'] = booking_delivery_melted['Service'].map({
    'has_table_booking': 'Table Booking',
    'has_online_delivery': 'Online Delivery'
})

# Plot
plt.figure(figsize=(12, 7))
sns.barplot(data=booking_delivery_melted, x='Count', y='city', hue='Service', palette='viridis',dodge=False,legend=False)
plt.title('Table Booking vs Online Delivery in Top Indian Cities', fontsize=16)
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```
